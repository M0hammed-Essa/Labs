```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the image

image = cv2.imread("images/cameraman.png", cv2.IMREAD_GRAYSCALE)
```

```python
# 1.	Increase the size of the image. 

resized_image = cv2.resize(image, (500, 500))
cv2.imshow("Cameraman1", image)
cv2.imshow("Cameraman2", resized_image)
cv2.waitKey()
```

```python
# 2.	Rotate the image by 120 degrees.

# Get the image dimensions
height, width = image.shape[:2]
 
# Define the rotation center
center = (width // 2, height // 2)

# Get the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, 190, 1.0)
 
# Perform the rotation
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
cv2.imshow("Cameraman", rotated_image)
cv2.waitKey()
```

```python
# 3.	Perform Sheer operations over the image.

image = cv2.imread("images/cameraman.png", cv2.IMREAD_GRAYSCALE)
height, width = image.shape[:2]

# Shear factor
shear_factor = 0.5

# Shear matrix (horizontal shear)
M = np.float32([[1, shear_factor, 0],
                [0, 1, 0]])

# Apply shear transformation
sheared_img = cv2.warpAffine(image, M, (int(width + shear_factor*height), height))
cv2.imshow("Cameraman", sheared_img)
cv2.waitKey()
```

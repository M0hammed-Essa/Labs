```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the image

image = cv2.imread("images/cameraman.png", cv2.IMREAD_GRAYSCALE)
```

```python
# 1 Negative Transformation

Negative_image = 255- image

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(Negative_image, cmap='gray')
plt.title("Negative")
plt.axis('off')

plt.show()
```

```python
# 2.Log Transformation

# scaling constant
c = 1.0
Log_image = c * np.log1p(image)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(Log_image, cmap='gray')
plt.title("Log")
plt.axis('off')

plt.show()
```

```python
# 3.Power law Transformation


gamma_image_Bright = np.power(image / 255.0, 0.1) * 255

gamma_image_Dark = np.power(image / 255.0, 5.0) * 255

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gamma_image_Bright, cmap='gray')
plt.title("Power law - Bright")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(gamma_image_Dark, cmap='gray')
plt.title("Power law - Dark ")
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage import data, img_as_float
from skimage import exposure
```

```python
image = data.moon()
```

```python
# Task 1

p3, p80 = np.percentile(image, (3, 80))
img_rescale = exposure.rescale_intensity(image, in_range=(p3, p80))

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_rescale, cmap='gray')
plt.title('Rescaled Image')
plt.axis('off')

plt.show()
```

```python
# Task 2

fig = plt.figure(figsize=(10, 7))
fig.add_subplot(2, 1, 1)
plt.imshow(image, cmap = 'gray')
plt.axis('off')
plt.title('Original image')


flat_image = exposure.equalize_hist(image)

fig.add_subplot(2, 1, 2)
plt.hist(flat_image.ravel(), bins=256, range=(0, 1))
plt.title("Histogram After Equalization")
plt.show()
```

```python
# Task 3

# Source image (will be modified)
source = data.chelsea()

# Reference image (histogram will be copied)
reference = data.rocket()

match = exposure.match_histograms(source, reference, channel_axis=-1)

plt.figure(figsize=(10,10))

plt.subplot(1,3,1)
plt.imshow(source, cmap='gray')
plt.title('Source (Chelsea)')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(reference, cmap='gray')
plt.title('Reference (Rocket)')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(match.astype(np.uint8))
plt.title('Matched Image')
plt.axis('off')

plt.show()
```
